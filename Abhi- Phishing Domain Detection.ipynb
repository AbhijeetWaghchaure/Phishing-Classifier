```python
shape=result.shape
shape
```

```python
#detecting and removing outlier in ttl_hostname
Q1=np.percentile(result['ttl_hostname'],25,interpolation='midpoint')
Q3=np.percentile(result['ttl_hostname'],75,interpolation='midpoint')
IQR=Q3-Q1
```

```python
print("Q1=", Q1)
print("Q3=", Q3)
print("IQR=", IQR)
```

```python
Q3+1.5*IQR
```

```python
upper=np.where(result.drop('phishing',axis=1) >= (Q3+1.5*IQR))
```

```python
lower=np.where(result.drop('phishing',axis=1) <= (Q1-1.5*IQR))
```

```python
result.drop(upper[1],inplace=True)
```

```python
result.drop(lower[0],inplace=True)
```

```python
new_shape=result.shape

print("Old shape : ",shape)
print("After Removing outlier ")
print("New shape : ",new_shape)
```

```python
# shuffling the rows in the dataset so that when splitting the train and test set are equally distributed
result = result.sample(frac=1).reset_index(drop=True)
result.head()
```

```python
# Sepratating & assigning features and target columns to X & y
y = result['phishing']
X = result.drop('phishing',axis=1)
X.shape, y.shape
```

```python
# Splitting the dataset into train and test sets: 80-20 split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 100)
X_train.shape, X_test.shape
```

```python
#importing packages
from sklearn.metrics import accuracy_score
```

```python
# Creating holders to store the model performance results
ML_Model = []
acc_train = []
acc_test = []

#function to call for storing the results
def storeResults(model, a,b):
    ML_Model.append(model)
    acc_train.append(round(a*100, 3))
    acc_test.append(round(b*100, 3))
```

# Decision Tree model

```python
# Decision Tree model 
from sklearn.tree import DecisionTreeClassifier

# instantiate the model 
tree = DecisionTreeClassifier(max_depth = 5)
# fit the model 
tree.fit(X_train, y_train)
```

```python
#predicting the target value from the model for the samples
y_test_tree = tree.predict(X_test)
y_train_tree = tree.predict(X_train)
```

```python
#computing the accuracy of the model performance
acc_train_tree = accuracy_score(y_train,y_train_tree)
acc_test_tree = accuracy_score(y_test,y_test_tree)

print("Decision Tree: Accuracy on training Data: {:.3f}".format(acc_train_tree))
print("Decision Tree: Accuracy on test Data: {:.3f}".format(acc_test_tree))
```

```python
#checking the feature improtance in the model
plt.figure(figsize=(30,30))
n_features = X_train.shape[1]
plt.barh(range(n_features), tree.feature_importances_, align='center')
plt.yticks(np.arange(n_features), X_train.columns)
plt.xlabel("Feature importance")
plt.ylabel("Feature")
plt.show()
```

```python
#y_test vs y_test_predict
print(sns.scatterplot(x=y_test, y=y_test_tree))
```

```python
#storing the results. The below mentioned order of parameter passing is important.
#Caution: Execute only once to avoid duplications.
storeResults('Decision Tree', acc_train_tree, acc_test_tree)
```
